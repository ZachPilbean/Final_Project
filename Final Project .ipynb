```python
import random
random=random.randint(1,20)

bucket8=0
bucket5=0
bucket3=0
bucket20=0

def display():
    print('8L bucket: ', bucket8)
    print('5L bucket: ', bucket5)
    print('3L bucket: ', bucket3)
    print('20L bucket:', bucket20)

active=True 
while active:
    
    choose_bucket= input(f"Which bucket do you want to use? (8L, 5L, 3L) Enter stop when you have {random} Liters in the  20L bucket. ")
    if choose_bucket=='8L':
        bucket_action1= input('Do you want to fill the bucket, pour the bucket, or empty the bucket? ')
        if bucket_action1=='fill':
            bucket8=8
            display()
        elif bucket_action1=='empty':
            bucket8=0
            display()
        elif bucket_action1=='pour':
            receive_bucket1= input('Which bucket do you want to pour it into? ')
            if receive_bucket1=='3L':
                bucket8=bucket8-(3-bucket3)
                bucket3=3 
                display()
            elif receive_bucket1=='5L':
                bucket8=bucket8-(5-bucket5)
                bucket5=5
                display()
            elif receive_bucket1=='20L':
                bucket20=bucket20+bucket8
                bucket8=0
                display()
                
                
    elif choose_bucket=='5L':
        bucket_action2=input('Do you want to fill the bucket, pour the bucket, or empty the bucket? ')
        if bucket_action2=='fill':
            bucket5=5
            display()
        elif bucket_action2=='empty':
            bucket5=0
            display()
        elif bucket_action2=='pour':
            receive_bucket2= input('Which bucket do you want to pour it into?  ')
            if receive_bucket2=='8L':
                bucket8=bucket8+bucket5
                bucket5=0
                display()
            elif receive_bucket2=='3L':
                bucket5=bucket5-(3-bucket3)
                bucket3=3
                display()
            elif receive_bucket2=='20L':
                bucket20=bucket20+bucket5
                bucket5=0
                display()
                
            
    elif choose_bucket=='3L':
        bucket_action3=input('Do you want to fill the bucket, pour the bucket, or empty the bucket? ')
        if bucket_action3=='fill':
            bucket3=3
            display()
        elif bucket_action3=='empty':
            bucket3=0
            display()
        elif bucket_action3=='pour':
            receive_bucket3=input('Which bucket do you want to pour it into? ')
            if receive_bucket3=='8L':
                bucket8=bucket8+bucket3
                bucket3=0  
                display()
            elif receive_bucket3=='5L':
                bucket5=bucket5+bucket3
                bucket3=0
                display()
            elif receive_bucket3=='20L':
                bucket20=bucket20+bucket3
                bucket3= 0
                display()
                
                
    elif choose_bucket=='stop':
        display()
        if bucket20==random:
            print('Congratulations! You won! ')
            active=False
        else:
            print('You didn\'t get it. Keep trying. ')
```
